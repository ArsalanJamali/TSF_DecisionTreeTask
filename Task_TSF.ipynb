## Import Libraries

```python
import pandas as pd
from sklearn.tree import plot_tree,DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

## Reading the dataset

```python
data=pd.read_csv("iris.csv",index_col=0)
```

```python
data.info()
```

## Feature Extraction

```python
X=data.iloc[:,0:4]
Y=data.iloc[:,4]
```

```python
X
```

```python
Y
```

## Splitting The Dataset into Training(80%) and Testing(*30*%) data

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,train_size=0.80,random_state=2)
```

```python
decision_tree=DecisionTreeClassifier(criterion="entropy")
```

## Training the model

```python
decision_tree.fit(xtrain,ytrain)
```

## The accuracy of the model

```python
score=decision_tree.score(xtest,ytest)
print("The accuracy of model is ",score)
```

## visualization of tree

```python
names=[]
names=data.columns
```

```python
fig=plt.figure(figsize=(20,15))
plot_tree(decision_tree,feature_names=names[[0,1,2,3]],class_names=names[4],filled=True)
```
